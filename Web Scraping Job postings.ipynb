This code for down below is for finding the first job posting on the website

```python
from bs4 import BeautifulSoup
import requests
import time
"imports complete"
```

```python
html_text = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=").text
soup = BeautifulSoup(html_text, "lxml")
#li stands for list
#when using this .find function, the first argument is called the html tag that we are searching for
all_jobs = soup.find_all("li", class_ = "clearfix job-bx wht-shd-bx")
```

```python
#only finds the first job with that tag
job = soup.find("li", class_ = "clearfix job-bx wht-shd-bx")

#finds the name of the company for the first job list posting with html code
#.text finds just the text of the company name
#.replace gets rid of spaces
company_name = job.find("h3", class_ = "joblist-comp-name").text.replace(" ", "")
print(company_name)
```

```python
#finds the skills listed inside the job posting
skills = job.find("span", class_ = "srp-skills").text.replace(" ","")
print(skills)
```

```python
#Write a nice printed statement
print(f'''
Company Name: {company_name}
Required Skills: {skills}
''')
```

```python
#there is a span inside of a span which is why we have to add the .span at the end of this statement
published_date = job.find("span", class_ = "sim-posted").span.text
print(published_date)
```

In order to find all the job postings for the website, just add _all

```python
all_jobs = soup.find_all("li", class_ = "clearfix job-bx wht-shd-bx")

#write a for loop to cycle through all of the job posting boxes
for job in all_jobs:
    company_name = job.find("h3", class_ = "joblist-comp-name").text.replace(" ", "")
    skills = job.find("span", class_ = "srp-skills").text.replace(" ","")
    published_date = job.find("span", class_ = "sim-posted").span.text
    
    print(f'''
    Company Name: {company_name}
    Required Skills: {skills}
    ''')
    
    print("")
```

Now we want to filter to just the jobs that say posted a few days ago

```python
for job in all_jobs:
    published_date = job.find("span", class_ = "sim-posted").span.text
    if "few" in published_date:
        company_name = job.find("h3", class_ = "joblist-comp-name").text.replace(" ", "")
        skills = job.find("span", class_ = "srp-skills").text.replace(" ","")
    
        print(f'''
        Company Name: {company_name}
        Required Skills: {skills}
        ''')
        
        print(published_date)
    
        print("")
```

Wrap our entire code in a function

```python
print("put some skill that you are not familiar with")
unfamiliar_skill = input(">")
print("filtering out {unfamiliar_skill}")

def find_jobs():
    html_text = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=").text
    soup = BeautifulSoup(html_text, "lxml")
    all_jobs = soup.find_all("li", class_ = "clearfix job-bx wht-shd-bx")

#enumerate allows to iterate over job list and job content itself

    for index, job in enumerate(all_jobs):
        published_date = job.find("span", class_ = "sim-posted").span.text
        if "few" in published_date:
            company_name = job.find("h3", class_ = "joblist-comp-name").text.replace(" ", "")
            skills = job.find("span", class_ = "srp-skills").text.replace(" ","")
            more_info = job.header.h2.a["href"]

#filters out unfamiliar skill
            if unfamiliar_skill not in skills:
                
#opens directory to write text file
#reads as such: with ___ open __ file and write __ as file
                with open(f"posts/{index}.txt", "w") as f:
                    f.write(f"Company Name: {company_name.strip()} \n")
                    f.write(f"Required Skills: {skills.strip()} \n")
                    f.write(f"More Info: {more_info} \n")
                print(f"File saved: {index}")
                
                
if __name__ == "__main__":
    while True:
        find_jobs()
        #allows your program to wait, it's in seconds
        time_wait_min = 10
        print(f"Waiting {time_wait_min} minutes...")
        time.sleep(time_wait_min * 60)
```
